# Task#1

```python
import pandas as pd
clicks1 = pd.read_csv('data/clicks/part-001.csv',index_col = 0)
clicks2 = pd.read_csv('data/clicks/part-002.csv',index_col = 0)
clicks3 = pd.read_csv('data/clicks/part-003.csv',index_col = 0)
clicks4 = pd.read_csv('data/clicks/part-004.csv',index_col = 0)
clicks4.pop('screen')
clicks = pd.concat([clicks1, clicks2, clicks3, clicks4])
clicks.reset_index()
```

```python
total_clicks = clicks.groupby('date')['user_id'].count().reset_index()
total_clicks.rename(columns={'user_id': 'count'})
```

```python
total_clicks.to_csv('data/total_clicks.csv')
```

# Task #2

```python
users1 = pd.read_csv('data/users/part-001.csv', index_col = 0)
users2 = pd.read_csv('data/users/part-002.csv', index_col = 0)
users2.pop('city')
users = pd.concat([users1, users2])
users.reset_index()
```

```python
filtered_clicks = clicks.merge(users, how='left', left_on='user_id', right_on ='id')
filtered_clicks[filtered_clicks['country'] == 'LT'].reset_index(drop=True)
```

```python
filtered_clicks[filtered_clicks['country'] == 'LT'].reset_index(drop=True).to_csv('data/filtered_clicks.csv')
```
